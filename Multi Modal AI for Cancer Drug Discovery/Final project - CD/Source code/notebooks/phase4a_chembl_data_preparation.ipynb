```python
from chembl_webresource_client.new_client import new_client
import pandas as pd

activity = new_client.activity

res = activity.filter(
    standard_type="IC50",
    standard_units="nM"
).only([
    "molecule_chembl_id",
    "canonical_smiles",
    "standard_value"
])[:5000]

data = list(res)
print("Records fetched:", len(data))
```

```python
df = pd.DataFrame(data)

# remove rows with missing values
df = df.dropna(subset=["canonical_smiles", "standard_value"])

print("After cleaning:", len(df))
df.head()
```

```python
df.to_csv(
    "../data/raw/chembl/chembl_ic50_5000.csv",
    index=False
)

print("Saved to data/raw/chembl/chembl_ic50_5000.csv")
```
