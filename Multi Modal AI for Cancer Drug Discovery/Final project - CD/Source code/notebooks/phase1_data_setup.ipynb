```python
from pathlib import Path

DATASET_ROOT = Path("E:/Final_Year_Project/data/raw/breakhis/BreaKHis_v1/histology_slides/breast")

print("Dataset path exists:", DATASET_ROOT.exists())
print("Subfolders:", [p.name for p in DATASET_ROOT.iterdir()])
```

```python
IMAGE_EXTENSIONS = (".png", ".jpg", ".jpeg", ".tif")

image_paths = []

for root, dirs, files in os.walk(DATASET_ROOT):
    for file in files:
        if file.lower().endswith(IMAGE_EXTENSIONS):
            image_paths.append(os.path.join(root, file))

print("Total images found:", len(image_paths))
```

```python
from collections import Counter

magnification_counter = Counter()

for path in image_paths:
    if "40X" in path:
        magnification_counter["40X"] += 1
    elif "100X" in path:
        magnification_counter["100X"] += 1
    elif "200X" in path:
        magnification_counter["200X"] += 1
    elif "400X" in path:
        magnification_counter["400X"] += 1

magnification_counter
```

```python
def show_sample_image(mag):
    for path in image_paths:
        if mag in path:
            img = Image.open(path)
            plt.imshow(img)
            plt.axis("off")
            plt.title(f"Sample image at {mag}")
            plt.show()
            break

for mag in ["40X", "100X", "200X", "400X"]:
    show_sample_image(mag)
```
